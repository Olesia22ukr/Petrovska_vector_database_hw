````python
import pandas as pd

# беремо перші 1000 статей
papers = dataset.select(range(1000))

df = pd.DataFrame({
    "id": range(len(papers)),
    "text": papers["abstract"]
})

df.head()
````

````python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

print("Embedding model OK")
````

````python
embedding = model.encode(
    df["text"][0]
)

print(len(embedding))
````

````python
embeddings = model.encode(
    df["text"].tolist(),
    show_progress_bar=True
)

print(embeddings.shape)
````

````python
records = []

for i in range(len(df)):
    records.append({
        "id": str(i),
        "values": embeddings[i].tolist(),
        "metadata": {
            "text": str(df["text"][i])[:1000]
        }
    })

print(records[0].keys())
````

````python
from pinecone import Pinecone

pc = Pinecone(
    api_key="my api"
)

print("Pinecone connected")
````

````python
from pinecone import ServerlessSpec

index_name = "arxiv-search"

if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=384,
        metric="cosine",
        spec=ServerlessSpec(
            cloud="aws",
            region="us-east-1"
        )
    )

print("Index created")
````

````python
index = pc.Index("arxiv-search")

print(index)
````

````python
batch_size = 100

for i in range(0, len(records), batch_size):
    batch = records[i:i + batch_size]

    index.upsert(
        vectors=batch
    )

print("Vectors uploaded")
````

````python
def fixed_chunk(text, chunk_size=300):
    return [
        text[i:i+chunk_size]
        for i in range(0, len(text), chunk_size)
    ]


fixed_chunks = []

for text in df["text"]:
    fixed_chunks.extend(
        fixed_chunk(text)
    )

print("Fixed chunks:", len(fixed_chunks))
````

````python
def overlapping_chunk(text, chunk_size=300, overlap=50):
    chunks = []

    start = 0

    while start < len(text):
        chunks.append(
            text[start:start+chunk_size]
        )
        start += chunk_size - overlap

    return chunks


semantic_chunks = []

for text in df["text"]:
    semantic_chunks.extend(
        overlapping_chunk(text)
    )

print("Semantic chunks:", len(semantic_chunks))
````

````python
from rank_bm25 import BM25Okapi
import numpy as np

documents = df["text"].tolist()

tokenized_documents = [
    doc.lower().split()
    for doc in documents
]

bm25 = BM25Okapi(tokenized_documents)

print("BM25 ready")
````

````python
query = "computer vision deep learning"

scores = bm25.get_scores(
    query.lower().split()
)

top_indices = np.argsort(scores)[-3:][::-1]

for idx in top_indices:
    print("BM25 score:", round(scores[idx], 3))
    print(df.iloc[idx]["text"][:500])
    print("=" * 50)
````

````python
query = "machine learning neural networks"

# Vector search
query_embedding = model.encode(query).tolist()

vector_results = index.query(
    vector=query_embedding,
    top_k=3,
    include_metadata=True
)

print("VECTOR SEARCH")
for r in vector_results["matches"]:
    print(round(r["score"], 3))
    print(r["metadata"]["text"][:200])
    print("-"*40)


# BM25 search
bm25_scores = bm25.get_scores(query.lower().split())

bm25_top = np.argsort(bm25_scores)[-3:][::-1]

print("\nBM25 SEARCH")

for idx in bm25_top:
    print(round(bm25_scores[idx], 3))
    print(df.iloc[idx]["text"][:200])
    print("-"*40)
````

````python
def rrf_rank(vector_results, bm25_scores, k=60):
    scores = {}

    # vector ranking
    for rank, item in enumerate(vector_results["matches"]):
        doc_id = int(item["id"])
        scores[doc_id] = scores.get(doc_id, 0) + 1 / (k + rank + 1)

    # bm25 ranking
    bm25_top = np.argsort(bm25_scores)[::-1][:3]

    for rank, doc_id in enumerate(bm25_top):
        scores[doc_id] = scores.get(doc_id, 0) + 1 / (k + rank + 1)

    return sorted(
        scores.items(),
        key=lambda x: x[1],
        reverse=True
    )


query = "machine learning neural networks"

# vector search
query_embedding = model.encode(query).tolist()

vector_results = index.query(
    vector=query_embedding,
    top_k=3,
    include_metadata=True
)

# BM25
bm25_scores = bm25.get_scores(query.lower().split())


hybrid_results = rrf_rank(
    vector_results,
    bm25_scores
)


print("HYBRID SEARCH")

for doc_id, score in hybrid_results:
    print("RRF score:", round(score, 4))
    print(df.iloc[doc_id]["text"][:300])
    print("="*50)
````

````python
queries = [
    "machine learning neural networks",
    "artificial intelligence machine learning",
    "computer vision deep learning"
]


for query in queries:

    print("\nQUERY:", query)
    print("="*60)

    # Vector
    query_embedding = model.encode(query).tolist()

    vector_results = index.query(
        vector=query_embedding,
        top_k=3,
        include_metadata=True
    )

    print("\nVECTOR SEARCH:")
    for r in vector_results["matches"][:3]:
        print(round(r["score"],3), r["metadata"]["text"][:100])


    # BM25
    bm25_scores = bm25.get_scores(query.lower().split())

    bm25_top = np.argsort(bm25_scores)[-3:][::-1]

    print("\nBM25 SEARCH:")
    for idx in bm25_top:
        print(round(bm25_scores[idx],3), df.iloc[idx]["text"][:100])


    # Hybrid
    hybrid = rrf_rank(
        vector_results,
        bm25_scores
    )

    print("\nHYBRID SEARCH:")
    for idx, score in hybrid[:3]:
        print(round(score,4), df.iloc[idx]["text"][:100])
````

# Project Overview

This project demonstrates the implementation of a document retrieval system using vector databases.

The project includes:

- arXiv scientific papers dataset processing;
- text embeddings generation using all-MiniLM-L6-v2 model;
- vector storage in Pinecone;

# Part 1. Dataset Processing and Model Selection

## Dataset

The project uses the arXiv scientific papers dataset containing scientific paper abstracts.

The dataset was selected because scientific documents require semantic understanding and context-based retrieval rather than only exact keyword matching.


## Embedding Model Selection

The project uses the `all-MiniLM-L6-v2` embedding model.

The model converts text documents into numerical vectors that represent semantic meaning.

The main advantages:

- compact vector representation;
- efficient embedding generation;
- suitable for semantic similarity search.


## Vector Database Selection

Pinecone was selected as the vector database.

Reasons for choosing Pinecone:

- fast similarity search;
- scalable vector storage;
- metadata support;
- suitable for Retrieval-Augmented Generation (RAG) systems.


# Part 2. Vector Database Index Creation

## Index Configuration

The Pinecone index was created with:

- Dimension: 384
- Metric: cosine similarity
- Number of uploaded vectors: 1000


## Metadata

Each vector contains metadata with the original document text:

```text
{
    "text": "original document text"
}
```


## Vector Uploading

Embeddings were uploaded to Pinecone using batch processing.

Batch uploading improves efficiency when working with multiple vectors.


# Part 3. Search Methods Implementation

Three retrieval approaches were implemented:

1. Vector Search
2. BM25 Keyword Search
3. Hybrid Search


# 1. Vector Search

Vector search converts the user query into an embedding and searches for the closest document vectors using cosine similarity.

Advantages:

- understands semantic meaning;
- finds related documents even when exact words are different;
- works well with scientific texts.


Example query:

```
machine learning neural networks
```


# 2. BM25 Keyword Search

BM25 is a keyword-based retrieval method.

It ranks documents using:

- term frequency;
- importance of words;
- document statistics.


Advantages:

- good for exact keyword matching;
- fast retrieval;
- does not require embeddings.


Limitations:

- does not understand semantic meaning;
- depends on matching words.


# 3. Hybrid Search

Hybrid search combines:

- semantic vector search;
- BM25 keyword search.


This approach uses advantages of both methods and provides more balanced retrieval results.


## Search Results Analysis

Vector Search:

- retrieves documents with similar semantic meaning;
- can find relevant documents without exact keyword matches.


BM25 Search:

- retrieves documents containing important query terms;
- works better for specific keyword queries.


Hybrid Search:

- combines semantic and keyword ranking;
- improves retrieval quality by using both approaches.


# Part 4. Chunking Strategies Comparison

Chunking is the process of splitting documents into smaller parts before retrieval.

Two chunking strategies were implemented and compared.


## Fixed-size Chunking

Fixed-size chunking splits text into equal-size parts.

Advantages:

- simple implementation;
- predictable chunk size;
- fast processing.


Disadvantages:

- may split related information;
- context can be partially lost.


## Overlapping Chunking

Overlapping chunking creates chunks with partially repeated text between neighboring parts.

Advantages:

- preserves more context;
- reduces information loss between chunks;
- improves retrieval continuity.


Disadvantages:

- creates more chunks;
- requires more storage.


## Chunking Comparison

| Strategy | Advantages | Disadvantages |
|---|---|---|
| Fixed-size chunking | Simple and efficient | Possible context loss |
| Overlapping chunking | Better context preservation | More chunks and processing |


# Part 5. Hybrid Search and RRF

Hybrid search was implemented by combining Vector Search and BM25 results.

To merge rankings from different retrieval methods, Reciprocal Rank Fusion (RRF) was used.


## Reciprocal Rank Fusion

RRF combines multiple ranked lists.

The score is calculated based on document position:

```
RRF_score = Σ 1 / (k + rank)
```

where:

- rank is the document position;
- k is a constant used for smoothing.

## Mathematical Analysis

Vector Search uses cosine similarity:

```
similarity = cosine(query_vector, document_vector)
```

Higher similarity means that document vectors are closer in semantic space.


BM25 uses term frequency and inverse document frequency:

- frequent important terms increase relevance;
- common words have lower influence.


RRF combines rankings from different retrieval methods:

```
RRF_score = Σ 1 / (k + rank)
```
